```python
import matplotlib.pyplot as plt
import pandas as pd # v. 1.7
import MySQLdb
import numpy as np
import time
import sys
import datetime 
from datetime import datetime
from sympy import *
from sympy.solvers import solve
from sympy.parsing.sympy_parser import parse_expr
from sklearn.metrics import mean_absolute_error,mean_squared_error
import statistics 
%matplotlib inline

conn = MySQLdb.connect("localhost","conn","hems")
cursor = conn.cursor()

query_expand = "SELECT * FROM thesis.predict_expanding1 WHERE Timestamp >= '%s'   AND TimeStamp <= '%s';"% ("2016-10-01 00:00:00","2016-10-31 23:59:59")
query_moving = "SELECT * FROM thesis.predict_moving1 WHERE Timestamp >= '%s'   AND TimeStamp <= '%s';"% ("2016-10-01 00:00:00","2016-10-31 23:59:59")
query = "SELECT Timestamp, total_watt FROM thesis.load WHERE Timestamp >= '%s'   AND TimeStamp <= '%s';"% ("2016-10-01 00:00:00","2016-10-31 23:59:59")

df_expand= pd.read_sql(query_expand , conn , index_col=['Timestamp'])
df_moving   = pd.read_sql(query_moving  , conn , index_col=['Timestamp'])
df = pd.read_sql(query, conn , index_col=['Timestamp'])

df['pred_expanding'] = df_expand.watt
df['pred_moving'] = df_moving.watt

fig, ax = plt.subplots(figsize=(14, 6))
ax.set_title('pred_moving VS pred_expanding  VS exactly_power')
df.plot(ax=ax)
plt.show()
```

```python
##### EXPANDING 
from sklearn.metrics import mean_absolute_error,mean_squared_error
from sklearn.metrics import mean_squared_error
from math import sqrt


df.APE_ex = (np.abs((df.total_watt-df_expand.watt)/df.total_watt))
MAPE_ex = np.mean(df.APE_ex)*100
# df.to_csv('out.csv')
a = df.total_watt.sum()
b = df_expand.watt.sum()
print a,b

print "EXPANDING ___________________________"
print "--- type 1 : each data"
print "% MEAN ABSOLUTE PERCENT ERROR  :",MAPE_ex, "%"
print " MEAN ABSOLUTE ERROR                    :",mean_absolute_error(df.total_watt,df_expand.watt)
print " MEAN SQUARE ERROR                        :",mean_squared_error (df.total_watt , df_expand.watt) 
print " ROOT MEAN SQUARE ERROR               :",sqrt(mean_squared_error(df.total_watt, df_expand.watt))

print "--- type 2 : whole data"
print "% MEAN ABSOLUTE PERCENT ERROR   :", np.mean( (np.abs((a-b)/a))*100), "%"
print " MEAN ABSOLUTE ERROR                     :", np.sum(np.absolute(a - b))
print " MEAN SQUARE ERROR                        :", np.mean((a-b)**2)
print " ROOT MEAN SQUARE ERROR               :", sqrt(np.mean((a-b)**2))
print "% Error                                              :",(abs(df.total_watt.sum()-df_expand.watt.sum()))/df.total_watt.sum(), "%" 

  
print "\n\n"


##### EXPANDING 
from sklearn.metrics import mean_absolute_error,mean_squared_error
df.APE_mo = (np.abs((df.total_watt-df_moving.watt)/df.total_watt))
MAPE_mo = np.mean(df.APE_mo)*100
# df.to_csv('out.csv')


a1 = df.total_watt.sum()
b1 = df_moving.watt.sum()
print a1,b1
print "MOVING ___________________________"
print "--- type 1 : each data"
print "% MEAN ABSOLUTE PERCENT ERROR  :",MAPE_mo, "%"
print " MEAN ABSOLUTE ERROR                    :",mean_absolute_error(df.total_watt,df_moving.watt)
print " MEAN SQUARE ERROR                        :",mean_squared_error (df.total_watt , df_moving.watt) 
print " ROOT MEAN SQUARE ERROR               :",sqrt(mean_squared_error(df.total_watt, df_moving.watt))

print "--- type 2 : whole data"
print "% MEAN ABSOLUTE PERCENT ERROR   :", np.mean( (np.abs((a1-b1)/a1)))*100, "%"
print " MEAN ABSOLUTE ERROR                     :", np.sum(np.absolute(a1 - b1))
print " MEAN SQUARE ERROR                        :", np.mean((a1-b1)**2)
print " ROOT MEAN SQUARE ERROR               :", sqrt(np.mean((a1-b1)**2))
print "% Error                                              :",(abs(df.total_watt.sum()-df_moving.watt.sum()))/df.total_watt.sum(), "%" 

```

```python
### EAPANDING
import statsmodels.api as sm
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
import pandas as pd # v. 1.7
import MySQLdb
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import mean_absolute_error,mean_squared_error
import statistics 
%matplotlib inline 
%matplotlib inline 

conn = MySQLdb.connect("localhost","conn","hems")
cursor = conn.cursor()

query_expand = "SELECT * FROM thesis.predict_expanding1 WHERE Timestamp >= '%s'   AND TimeStamp <= '%s';"% ("2016-10-01 00:00:00","2016-10-31 23:59:59")
df_expand = pd.read_sql(query_expand , conn , index_col=['Timestamp'])

query = "SELECT Timestamp, total_watt FROM thesis.load WHERE Timestamp >= '%s'   AND TimeStamp <= '%s';"% ("2016-10-01 00:00:00","2016-10-31 23:59:59")
df = pd.read_sql(query, conn , index_col=['Timestamp'])

X = df_expand.watt
Y = df.total_watt

results = sm.OLS(Y,sm.add_constant(X)).fit()
print results.summary()
plt.figure(figsize=(14,6))
plt.scatter(X,Y)
X_plot = np.linspace(0,2400,2400)
# plt.plot(X_plot, X_plot*results.params[0] + results.params[1])
plt.plot(X_plot)
plt.title('Error of  comparing exactly data and predicted data (Expanding) in 1 month',fontsize=18)
plt.xlabel('Exactly data (watt)',fontsize=18)
plt.ylabel('Predicted data (watt)',fontsize=18)
# plt.show()
plt.ylim((-0.5,1200))
plt.xlim((-0.5,1200))


df.total_watt['prediction'] = df_expand.watt
df.plot

```
