```python
import pandas as pd
import MySQLdb
import numpy as np
import time
import datetime 
import random
import matplotlib.pyplot as plt
%matplotlib inline 
import statistics 
import requests
from sympy import *
from sympy.solvers import solve
from sympy.parsing.sympy_parser import parse_expr
# ____________________________________________________________________________________________________   

'''
##  PRIORITY TABLE
'''
class ListTable(list):
    """ Overridden list class which takes a 2-dimensional list of 
        the form [[1,2,3],[4,5,6]], and renders an HTML Table in 
        IPython Notebook. """
    
    def _repr_html_(self):
        html = ["<table>"]
        for row in self:
            html.append("<tr>")
            
            for col in row:
                html.append("<td>{0}</td>".format(col))
            
            html.append("</tr>")
        html.append("</table>")
      
        return ''.join(html)
 # ____________________________________________________________________________________________________   
    
"""
## CONVERSING  UNIT (kW) >>  BILL (bath)
"""
def bill(allpow,Ft):
    
    t1 = 3.2484      # rate_1
    t2 = 4.2218      # rate_2
    t3 = 4.4217      # rate_3
    ser = 38.22      # service_charge
 
    ## Fundamental parameter      
    if allpow <= 0:
        bath = 0
    elif 0 <= allpow <= 150:
        bath = allpow*t1
    elif  151 <= allpow <= 400:
        power = allpow-150
        bath_1 = 150*t1
        bath_2 = power*t2
        bath = bath_1+bath_2
    else:
        power = allpow-150
        bath_1 = 150*t1
        power = power-250
        bath_2 = 250*t2
        bath_3 = power*t3
        bath = bath_1+bath_2+bath_3   
    service_price = 38.22  #permonth (fixed)   
    ## FT -33.2 satang/unit
    FT_price = allpow*Ft
    ## VAT 7%
    VAT_price = ((service_price+bath)+FT_price)*0.07
    bill = bath+service_price+FT_price+VAT_price
    return round(bill,2)
 # ____________________________________________________________________________________________________   

"""
## CONVERSING  BILL (bath) >>  UNIT (kW) >> 
"""
def inbill(bath,Ft):
    
    t1 = 3.2484      # rate_1
    t2 = 4.2218      # rate_2
    t3 = 4.4217      # rate_3
    ser = 38.22      # service_charge  
    u = symbols('u')
    test_1 = solve(Eq((u*t1)+ser+(u*Ft)+(0.07*((u*t1)+ser))+(0.07*(u*Ft)),bath))
    test_2 = solve(Eq((150*t1)+((u-150)*t2)+ser+(u*Ft)+(0.07*((150*t1)+((u-150)*t2)+ser))+(0.07*(u*Ft)),bath))
    test_3 = solve(Eq((150*t1)+(250*t2)+((u-400)*t3)+ser+(u*Ft)+(0.07*((150*t1)+(250*t2)+((u-400)*t3)+ser))+(0.07*(u*Ft)),bath))
    if 1 <=  test_1[0] <= 150:
        unit = test_1[0]
    elif 151 <=  test_2[0] <=  400:
        unit = test_2[0]
    else:
        unit = test_3[0]
    return (round(unit))
 # ____________________________________________________________________________________________________   

"""
## CHECKING  The ENDING POSITION OF EACH APPLIANCES
"""
def end_air(j): 
    check = 1  
    m_air = j+1   
    while (check == 1):
        while df1.aircont[ m_air] > 0 and m_air < (96*182)-1:
            m_air += 1
        check = 0
    return  m_air

def end_com(j): 
    check = 1  
    m_com = j+1   
    while (check == 1):
        while df1.computer[m_com] > 0 and m_com < (96*182)-1:
            m_com += 1  
        check = 0
    return  m_com

def end_elec(j): 
    check = 1  
    m_elec = j+1   
    while (check == 1):
        while df1.elec_thermo[m_elec] > 0 and m_elec < (96*182)-1:
            m_elec += 1  
        check = 0
    return  m_elec

def end_hair(j): 
    check = 1  
    m_hair = j+1   
    while (check == 1):
        while df1.hair_dryer[m_hair] > 0 and m_hair < (96*182)-1:
            m_hair += 1  
        check = 0
    return  m_hair

def end_iphon(j): 
    check = 1  
    m_iphon = j+1   
    while (check == 1):
        while df1.iphonecharger[ m_iphon] > 0 and m_iphon < (96*182)-1:
            m_iphon += 1  
        check = 0
    return  m_iphon

def end_fan(j): 
    check = 1  
    m_fan = j+1   
    while (check == 1):
        while df1.fan[m_fan] > 0 and m_fan < (96*182)-1:
            m_fan += 1  
        check = 0
    return  m_fan

def end_tv(j): 
    check = 1  
    m_tv = j+1   
    while (check == 1):
        while df1.tv[m_tv] > 0 and m_tv < (96*182)-1 :
            m_tv += 1  
        check = 0
    return  m_tv

def end_water(j): 
    check = 1  
    m_water = j+1   
    while (check == 1):
        while df1.water_heater[m_water] > 0 and m_water < (96*182)-1:
            m_water += 1  
        check = 0
    return  m_water
 # ____________________________________________________________________________________________________   

"""
## CHECKING  The STATE OF APPLIANCES 
"""    
def state(j,m_air,m_com, m_elec,m_hair,m_iphon, m_fan ,m_tv, m_water) :
    if j <= m_air:
        ans1 = 18    
    else:
        ans1 = 0
    if j <= m_com:
        ans2 = df1.computer[j]
    else:
        ans2 = 0
    if j <= m_elec:
        ans3 = df1.elec_thermo[j]
    else:
        ans3 = 0
    if j <= m_hair:
        ans4 = df1.hair_dryer[j]
    else:
        ans4 = 0
    if j <= m_iphon:
        ans5 = df1.iphonecharger[j]
    else:
        ans5 = 0
    if j <= m_fan:
        ans6 = df1.fan[j]
    else:
        ans6 = 0
    if j <= m_tv:
        ans7 = df1.tv[j]
    else:
        ans7 = 0
    if j <= m_water:
        ans8 = df1.water_heater[j]
    else:
        ans8 = 0

    k = ans1+ans2+ans3+ans4+ans5+ans6+ans7+ans8
#     print "substact k watt out from watt = ",k,"watt    ||||||" ,ans1,ans2,ans3,ans4,ans5,ans6,ans7,ans8
    return (ans1,ans2,ans3,ans4,ans5,ans6,ans7,ans8,k) 
 # ____________________________________________________________________________________________________   

"""
## PREDICTION 
"""        
def initial_trend(series, slen):
    sum = 0.0
    for i in range(slen):
        sum += float(series[i+slen] - series[i]) / slen
    return sum / slen

def triple_exponential_smoothing(series, slen, alpha, beta, gamma, n_preds):
    result = []
    seasonals = initial_seasonal_components(series, slen)
    for i in range(len(series)+n_preds):
        if i == 0: # initial values
            smooth = series[0]
            trend = initial_trend(series, slen)
            result.append(series[0])
            continue
        if i >= len(series): # we are forecasting
            m = i - len(series) + 1
            result.append((smooth + m*trend) + seasonals[i%slen])
        else:
            val = series[i]
            last_smooth, smooth = smooth, alpha*(val-seasonals[i%slen]) + (1-alpha)*(smooth+trend)
            trend = beta * (smooth-last_smooth) + (1-beta)*trend
            seasonals[i%slen] = gamma*(val-smooth) + (1-gamma)*seasonals[i%slen]
            result.append(smooth+trend+seasonals[i%slen])
    return result

def initial_seasonal_components(series, slen):
    seasonals = {}
    season_averages = []
    n_seasons = int(len(series)/slen)
    # compute season averages
    for j in range(n_seasons):
        season_averages.append(sum(series[slen*j:slen*j+slen])/float(slen))
    # compute initial values
    for i in range(slen):
        sum_of_vals_over_avg = 0.0
        for j in range(n_seasons):
            sum_of_vals_over_avg += series[slen*j+i]-season_averages[j]
        seasonals[i] = sum_of_vals_over_avg/n_seasons
    return seasonals
 # ____________________________________________________________________________________________________   

"""
## FITTING PARAMETERS FOR PREDICTION
""" 
from sklearn.metrics import mean_squared_error
def fittingparameter(data_y,j):
    count_alpha = 0
    count_beta = 0
    count_gamma = 0
    alpha = float(0)
    beta = float(0)
    gamma = float(0)
    index_alpha =[]
    index_beta =[]
    index_gamma = []
    act = j
    data = data_y
    pred_no = 1
    ramda = 0.1
    MSE_holtload = []    
    ##############################################################
    ### ?????????????????????????????????????????????????????????? DATABASE
    query_last = "SELECT Timestamp,total_watt FROM thesis.Prac_device26 WHERE Timestamp = '%s' ;"% (time_b_last[act])
    print query_last
    df1 = pd.read_sql(query_last, conn)
    while (alpha <= float(1)):
        count_alpha += 1
        beta = float(0)
        while (beta <=float(1)):
            count_beta += 1 
            gamma = float(0)
            while (gamma <= float(1)):
                count_gamma += 1
                pred_h = triple_exponential_smoothing(data,96*7,alpha,beta,gamma,pred_no)
                pred_data = pred_h[len(pred_h)-(pred_no):]                                               
                MSE = mean_squared_error (df1.total_watt, pred_data)   
#                 print MSE              
                MSE_holtload.append(MSE)
                index_alpha.append(alpha)
                index_beta.append(beta)
                index_gamma.append(gamma)     
#                 print index_alpha,index_beta,index_gamma
                gamma += ramda
            beta += ramda
        alpha += ramda  
#     print  count_gamma,count_beta,count_alpha    
    index_MSE = MSE_holtload.index(min(MSE_holtload))
    MSE_min = MSE_holtload[index_MSE]
    alpha = index_alpha[index_MSE]
    beta = index_beta[index_MSE]
    gamma= index_gamma[index_MSE]
#     print "the minimum MSE of prediction =",MSE_min,"by alpha =",alpha,"beta =",beta,"gamma =",gamma
    return (alpha,beta,gamma)
 # ____________________________________________________________________________________________________   

"""
## CHECKING THE STATUS OF APPLIANCES ON THAT TIME
""" 
def appliances(j):
    ############################################################## (day + table_database)
    ### ??????????????????????????????????????????????????????????
    time_b = pd.date_range('2016-10-01 00:14:59', periods=96*d, freq='15T')
    conn = MySQLdb.connect("localhost","conn","hems")
    c = conn.cursor()
    query1 = "SELECT * FROM thesis.load WHERE  Timestamp = '%s'; "% (time_b[j])
#     print "Check status of appliances from"
#     print query1
    df1 = pd.read_sql(query1, conn) 
    
    for row in df1.iterrows():
        index, data = row
        state = data.tolist()
    i = 1
    status = []
    while(i < len(state)-1):
        dict = { 1:'refrigerator',2:'wireless_router',3:'aircont',4:'computer',5:'electrical_thermo',6:'hair_dryer',7:'microwave',8:'iphone_charger',9:'fan',10:'tv',
                11:'washing_machine',12:'iron',13:'rice_cooker',14:'water_heater',15:'light_system'}
        if state[i] != 0:
            a = dict[i]
            status.append(a)
        i +=1  
#     print "NOW !! you are turning ON" ,status ,"=",len(status),"items"
#     print "______________________________________________________________________________________________\n"
    return(state)
 # ____________________________________________________________________________________________________   
    
def notify(msg):
    URL = "https://notify-api.line.me/api/notify"
    ACCESS_TOKEN = "vxm278yAqFQcXbo2DlTbLD7pqO0QWU2McdNbiwhJWwp"
    MESSAGE_FIELD = {"message" : msg}
    LINE_HEADERS = { "Authorization": "Bearer " + ACCESS_TOKEN }
    try:
        response = requests.post( url=URL,headers=LINE_HEADERS,data=MESSAGE_FIELD)
#         print("Response HTTP Status Code: {status_code}".format( status_code=response.status_code))
    except requests.exceptions.RequestException:
        print("HTTP Request failed")  
```
